```python
data_with_100 = list()
for d in data:
    if not d['price'] == None and d['price']<= 100: 
        data_with_100.append(d)
print (data_with_100[0])
```

```python

country_price = defaultdict(list)
country_price_mean = defaultdict(int)
country_price_sdv = defaultdict(int)
for d in data_with_100:
    country_price[d['country']].append(d['price'])
for c in country_price:
    country_price_mean[c] = float(sum(country_price[c]))/len(country_price[c])
for c in country_price:
    country_price_sdv[c] = (float(sum([(p- country_price_mean[c])**2 for p in country_price[c]]))/len(country_price[c]))**(0.5)
    
```

```python
from sklearn.metrics import mean_squared_error

def getyear(d):
    n = d['title']
    for ns in n.split():
        if len(ns)== 4 and ns[:4].isdigit():
            return int(ns)
    return 0

#def winaryhash(d):

def feat(d):
    y = getyear(d)
    m  = country_price_mean[d['country']] 
    sdv = country_price_sdv[d['country']]
    feat = [1,y,m,sdv]
    #feat = [1,y,m]
    return feat
   
x_train = [feat(d) for d in data_with_100[:len(data_with_100)/2]]
y_train = list()
x_test = [feat(d) for d in data_with_100[len(data_with_100)/2:]]
y_test = list()
for d in data_with_100[:len(data_with_100)/2]:
    
    if  0 < d['price'] and  d['price'] <= 20:
        y_train.append(1)    
    elif  20 < d['price'] and  d['price'] <= 40:
        y_train.append(2)
    elif  40 < d['price'] and  d['price'] <= 60:
        y_train.append(3)
    elif  60 < d['price'] and  d['price'] <= 80:
        y_train.append(4)
    elif  80 < d['price'] and  d['price'] <= 100:
        y_train.append(5)

for d in data_with_100[len(data_with_100)/2:]:
    
    if  0 < d['price'] and  d['price'] <= 20:
        y_test.append(1)    
    elif  20 < d['price'] and  d['price'] <= 40:
        y_test.append(2)
    elif  40 < d['price'] and  d['price'] <= 60:
        y_test.append(3)
    elif  60 < d['price'] and  d['price'] <= 80:
        y_test.append(4)
    elif  80 < d['price'] and  d['price'] <= 100:
        y_test.append(5)
    
lin_clf = svm.LinearSVC(C=1,multi_class='ovr')
lin_clf.fit(x_train, y_train)
y_pre = lin_clf.predict(x_test)

correct = [  y_pre[i]==y_test[i]  for i in range(len(y_test))]
print'mse:', mean_squared_error(y_pre,y_test)
print 'accur:', float(sum(correct))/len(y_test)
```

```python
country_price = defaultdict(list)
country_price_mean = defaultdict(int)
country_price_sdv = defaultdict(int)

taster_price = defaultdict(list)
taster_price_mean = defaultdict(int)
taster_price_sdv = defaultdict(int)

winery_price = defaultdict(list)
winery_price_mean = defaultdict(int)
winery_price_sdv = defaultdict(int)

pt_price = defaultdict(list)
pt_price_mean = defaultdict(int)
pt_price_sdv = defaultdict(int)

varit_price = defaultdict(list)
varit_price_mean = defaultdict(int)
varit_price_sdv = defaultdict(int)

des_price = defaultdict(list)
des_price_mean = defaultdict(int)
des_price_sdv = defaultdict(int)

for d in data_with_100:
    country_price[d['country']].append(d['price'])
    taster_price[d['taster_name']].append(d['price'])
    winery_price[d['winery']].append(d['price'])
    pt_price[d['points']].append(d['price'])
    varit_price[d['variety']].append(d['price'])
    des_price[d['designation']].append(d['price'])

for des in des_price:
    des_price_mean[des] = float(sum(des_price[des]))/len(des_price[des])
for des in des_price:
    des_price_sdv[des] = (float(sum([(p - des_price_mean[des])**2 for p in des_price[des]]))/len(des_price[des]))**(0.5)     
    
for var in varit_price:
    varit_price_mean[var] = float(sum(varit_price[var]))/len(varit_price[var])
for var in varit_price:
    varit_price_sdv[var] = (float(sum([(p - varit_price_mean[var])**2 for p in varit_price[var]]))/len(varit_price[var]))**(0.5)    
    
for p in pt_price:
    pt_price_mean[p] = float(sum(pt_price[p]))/len(pt_price[p])
for pt in pt_price:
    pt_price_sdv[pt] = (float(sum([(p - pt_price_mean[pt])**2 for p in pt_price[pt]]))/len(pt_price[pt]))**(0.5)
    
for t in taster_price:
    taster_price_mean[t] = float(sum(taster_price[t]))/len(taster_price[t])
for t in taster_price:
    taster_price_sdv[t] = (float(sum([(p - taster_price_mean[t])**2 for p in taster_price[t]]))/len(taster_price[t]))**(0.5)

for w in winery_price:
    winery_price_mean[w] = float(sum(winery_price[w]))/len(winery_price[w])
for w in winery_price:
    winery_price_sdv[w] = (float(sum([(p - winery_price_mean[w])**2 for p in winery_price[w]]))/len(winery_price[w]))**(0.5)
    
for c in country_price:
    country_price_mean[c] = float(sum(country_price[c]))/len(country_price[c])   
for c in country_price:
    country_price_sdv[c] = (float(sum([(p- country_price_mean[c])**2 for p in country_price[c]]))/len(country_price[c]))**(0.5)
```
